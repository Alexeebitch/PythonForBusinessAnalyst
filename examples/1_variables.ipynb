 ## Переменные
Переменная— это именованная область памяти, используемая для хранения данных, которые могут изменяться в процессе выполнения программы.
Переменные позволяют программам сохранять и манипулировать информацией, такой как числа, строки текста или логические значения. Каждая переменная имеет имя, тип данных и значение. Имя используется для обращения к переменной, тип данных определяет вид информации, которую переменная может хранить, а значение — это текущие данные, сохраненные в переменной.

```python
a = 10  #<- теперь a - это переменная, в которой хранится значение 10
print(a)  #выведем переменную a, что бы убедиться что там лежит 10
```

Разберем более подробно написанный код. <br>
Для того что бы объявить переменную в перую очередь необходимо дать ей название. В самом начале строки мы сделали это. Назвали переменную "a".<br>
Далее написан оператор "=", который означает присваивание. После оператора присваивания обязательно должен указываться объект, который мы хотим присвоить переменной, в нашем случае - это число 10. <br>
Во второй строке мы вызываем функции "print" (о функциях будет рассказано позднее). У данной функции одна задача - получить объект и вывести его. В нашем случае мы передали в функцию переменную "a", после чего увидели значение, которое мы ранее записали в эту переменную.

## Типы данных
Типом данных или просто типом называют множество (допустимых) значений и совокупность операций над этими значениями. <br>
Компьютер не видит разницы между числом, строкой, датой - для него всё это лишь последовательности нулей и единиц. <br>
Тип данных подсказывает компьютеру, как следует обращаться с этими данными, предоставляя (ограничивая) операции для работы с данными этого типа.

Типы данных в Python можно услвно разделить на 2 основных типа:
1) Встроенные типы данных - это такие типы данных, которые уже определны внутри самого языка программирования;
2) Кастомные типы данных - это типы данных, которые разработаны в рамках устанавливаемых библиотек (о библиотеках будет расказано позже).

 Встроенные типы данных
Всего в Python можно выделить 10 основных типов данных, которые в свою очередь можно объеденить в группы:
1)  Числовые типы:<br>
     1.1) int - целые числа;<br>
     1.2) float - числа со знаками после запятой;<br>
     1.3) complex - комплексные числа.<br>
2)  Массивы:<br>
     2.1) str (строка) - набор состоящий из знаков (букв);<br>
     2.2) tuple (кортеж) - набор любых значений, неизменяемый;<br>
     2.3) list (список) - набор любых значений, изменяемый;<br>
3)  Множества:<br>
     3.1) set (множество) - набор уникальных значений.<br>
4)  Пустые типы:<br>
     4.1) None (ничего) - пустой тип данных.<br>
5)  Карты: <br>
     5.1) dict (словарь) - набор именованых элементов.<br>
6)  Логические типы: <br>
     6.1) bool (логический) - значение, которое может содержать одно из двух возможных значений 0 либо 1.<br>

#### Числовые типы данных
Числовые типы данных - это такие типы, которые представлены в виде числа. С ними можно производить любые арифметические операции. 

Математические операции в Python обозначаются следующими операторами:<br>
"+"  сложение; <br>
"-" вычитание; <br>
"/" деление; <br>
"*" умножение;<br>
"**" возведение в степень;<br>
"//" деление на цело; <br>
"%" остаток от деления. <br>

```python
a = 10 # <- так задается целочисленный тип
a = 10.1 # <- так задается тип данных с дробной частью
a = 1 + 1j  # <- так задается комлексный тип данных
```

```python
a = 10
b = 5
c = a + b 
d = a - b
e = a/b
f = a * b
g = a ** b
h = a//b
i = a % b
# Вывод результатов
print(f'Результат сложения {a} и {b} - это {c}')
print(f'Результат вычитания {b} из {a} - это {d}')
print(f'Результат деления {a} на {b} - это {e}')
print(f'Результат умножения {a} на {b} - это {f}')
print(f'Результат {a} в степени {b} - это {g}')
print(f'Результат целочисленного деления {a} на {b} - это {h}')
print(f'остаток от деления {a} на {b} - это {i}')
```

В данном блоке мы произвели основные математические операции над переменными a и b, при этом помещали результат каждого вычисления в разные переменные. После чего вывели все эти переменные, дополнив комментариями. Можете попробовать самостоятельно заменить значение переменных a и b и запустить код, при этом все сработает.

```python
a = 9.0423032423
b = 3.93
c = a + b 
d = a - b
e = a/b
f = a * b
g = a ** b
h = a//b
i = a % b
# Вывод результатов
print(f'Результат сложения {a} и {b} - это {c}')
print(f'Результат вычитания {b} из {a} - это {d}')
print(f'Результат деления {a} на {b} - это {e}')
print(f'Результат умножения {a} на {b} - это {f}')
print(f'Результат {a} в степени {b} - это {g}')
print(f'Результат целочисленного деления {a} на {b} - это {h}')
print(f'остаток от деления {a} на {b} - это {i}')
```

В данном блоке мы сменили тип данных на float, стоит обратить внимание на то, что изменился и тип возвращаемого значения. Во всех операциях, кроме деления, деление всегда возвращает тип float. Если вы проводите любую из перечисленных операций с разными типами, например a - int, b - float, то результат выполнения всегда будет float.

### Массивы
Массивы - это такие типы данных, которые способны хранить в себе множество упорядоченных значений.

#### <center>Списки</center>
Списки - это тип данных, который может хранить в себе коллекцию значений любого типа. А так же обладает свойствами добавления и удаления значения из списка.

```python
a = [] # <- так задается пустой список
a = [1, 2, 3] # <- так задается список целочисленных значений
a = [1.1, 2.2, 3.3] # <- так задается список значений float
a = [1, 2.2, [1,2,3]] # <- так тоже можно задать список, который не привязаан к типу данных
a = [[1,2,4],[2,3,5],[6,3,4]] # <- а так получается подобие матрицы
```

Что показывает приведенный код?<br>
Из этого кода мы можем понять, что в списке можно хранить все, что угодно.

```python
a = [1,10,2,100,2]
print(2 * a)
```

А таким образом можно продублировать все значения в списке. <br>
Но это скорее приятный бонус питона, но не часто применямая практика, об основных операциях со списком поговорим далее.


##### <center>Индексация и срезы</center>
Индексация - процесс обращения к элементу массива по индеку.<br>
Срезы - процесс, который позволяет получить подмассив из массива, путем указания индексов элементов.<br>

```python
first = a[0] # <- таким образом можно получить первый элемент списка, ВНИМАНИЕ: ИНДЕКСАЦИЯ В ПИТОНЕ НАЧИНАЕТСЯ С НУЛЯ
print(first)
```

```python
# ВНИМАНИЕ! МАГИЯ ПИТОНА
last = a[-1] # <- так можно получить последний элемент списка
print(last)
```

```python
b = a[0:2] # <- это срез, так можно получить первые 2 элемента массива
#(СРЕЗ БЕРЕТСЯ ОТ 1 УКАЗАННОГО ИНДЕНСА ДО ПОСЛЕДНЕГО, ПРИ ЭТОМ ПЕРВЫЙ УКАЗАННЫЙ ИНДЕС ВКЛЮЧАЕТСЯ В СРЕЗ, А ВТОРОЙ НЕ ВКЛЮЧАЕТСЯ)
print(b)
```

<center>
Легко запомнить как работают срезы, если вспомнить школьную программу и прямые с выколотыми точками.

##### Другие операции со списком

```python
c = [1,2,3]
d = a + c # <- так можно складывать списки, эта операция накзывается КОНКАТЕНАЦИЯ
print(d)
```

```python
a = [1,2,3]
print(a)
a.append(2) # <- таким образом можно добавить объект в конец списка, тем самым увелчить его длинну
print(a)
```

```python
a.count(1) # <- таким способом можно узнать количество конкретных элементов в списке
```

```python
a.reverse() # <- так можно перевернуть значения в списке
print(a)
```

```python
a.sort() #  <- так можно отсортировать значения в списке
print(a)
```

```python
a.insert(1,10) # <- так можно вставить значение в список, сперва указывается индекс, затем значение, которое необходимо вставить
print(a)
```

```python
a.pop(1) # <- так можно убрать значение из списка по индексу
print(a)
```

```python
a.remove(2) # <- так можно убрать указанные значения из списка (если таких значений несколько, уберется только одно - первое)
print(a)
```

```python
a.clear()
print(a) # <- таким способом можно очистить список, так же это можно сделать задав a пустым, использовав оператор присваивания
```

#### <center>Строки</center>
Списки - это как список, который хранит в себе букввы, но не совсем.

```python
stroka = "Тут текст" # <- так задается строка
stroka = 'Тут текст' # <- так тоже задается строка
```

Почему строки могут задаваться как в одинарных, так и в двойных ковычках?<br>
Ответ достаточно прост - прдставьте что вам, в вашем текста необходимо вставить цитату.<br>
Питон позаботился об этом, решение такой задачи показано ниже.

```python
stroka = 'Цитируя автора: "тут цитата"'
print(stroka)
```

Так как строки были нами ранее отнесены к группе "Массивы", логично, что основной фукнционал массивов, а именно - индексация и срезы у строк должны присутствовать.

```python
first = stroka[0] # <- первый элемент строки
print(first)
```

```python
b = stroka[0:3] # <- срез
print(b)
```

Какие еще есть свойства у строки мы разберем ниже

```python
b = stroka + stroka # <- строки можно складывать
print(b)
```

```python
b = stroka *10 # строки можно умножать на целое число
print(b)
```

```python
stroka = 'ЦиТирУя автора: "тУт цитата"'
b = stroka.capitalize() # <-строки можно приводить к привычному виду
print(b)
```

```python
print(stroka.upper())# можно сделать все буквы заглавными
print(stroka.lower())# или сделатб все прописными
```

```python
print(stroka.replace('а','б')) # в строке можно подменять буквы
print(stroka.replace('та','ра')) # или даже части строки
```

```python
print(stroka.split(' ')) # таким образом можно разделить строку по пробелам
print(stroka.split('т')) # и даже по буквам
```

!! Обратите внимание, что в случае деления строки мы получаем массив из строк<br>
А так же символ для разделения отсутствует в этом списке.

### Ничего
Ничего или None - это тип данных, с которым связано множество проблем при разработке.<br>
С ним нельзя проводить арифметические операции, однако есть некоторые примеры, когда None может быть полезен, такие примеры мы разберем в следующих этапах применения питона.<br>

```python
a = None # <- так задается пустота
print(a)
```

### Логика(bool)
Почему логикой назвали тип данных, который может принимать только 2 значения True(истина) или False(ложь)?<br>
Такое название скорее всего пришло из компьютерной логики, ведь все операции в конечном итоге базируются на нулях и единицах.

```python
a = True # <- так задается истина
b = False # <- a так ложь
```

C логическими типами есть несколько операции, которые пригодятся в дальнейшем.<br>
В Python есть 3 основных логических оператора:<br>
1) not - отрицание, оно способно превратить истину в ложь или ложь в истину;<br>
2) or - логическое "или", возвращает истину, если хотя бы одно из выражений является истиной;<br>
3) and - логическое "и" возвращает ложь если хотя бы одно из условий ложно.


```python
print(not a) # так можно превратить истину в ложь или ложь в истину, это называется ИНВЕРСИЯ
```

```python
print(a or b) # <- логическое или 
```

```python
print(a and b) # <- логическое и 
```

### Карты(Map)
Карты - это такая структура, которая хранит в себе записи в формате ключ - значение.<br>
Карты позволяют производить быстрый поиск по ключу. Трудно дать емкое определение тому, что такое HashMap, с нуля, легче поспринимать его в примерах<br>

#### <center>Словарь</center>
<center>В питоне есть только один встроенный HashMap - он называется словарь.</center>

```python
contacts = {} # <- так задается пустой словарь
```

Наверное самый частый пример объяснения словаря - это составление телефонной книги.

```python
contacts = {"Игорь": "+79874835901","Иван": "+79348510324"} 
print(contacts)
```

Разбирая ранее написанный код, можно сказать следующее:
1) Элементы словаря задаются так же как в списке, через запятую;
2) Каждый элемент словаря задается как пара ключ : значение;

Для чего это нужно?<br>
Давайте обратимя к элементу словаря

```python
contacts['Игорь']
```

Обращение в словаре похоже на индексацию, не правда ли?<br>
Однако есть существенное отличие, вместо индекса здесь используется ключ. Это может быть удобно, когда мы хотим хранить именованные объекты. Как бы мы действовали, если бы не знали о словарях?

```python
# Скорее всего мы бы сделали 2 списка
names = ['Игорь', 'Иван'] # <- список с именами
tels = ['+79874835901',"+79348510324"] # <- список с телефонами
idx = names.index('Игорь') # <- Находили бы индекс по имени
tels[idx] # <- после чего обращались бы к другому списку по найденному индексу
```

Но, если бы у нас в списке было 2 Ивана, получилось бы, что мы не смолги заведомо точно узнать какого именно Ивана мы ищем.<br>
Эти проблемы решает словарь:
1) Ключи должны быть в единственном экземпляре, не может быть 2 'Иванов в одной телефонной книге';
2) При обращении по ключу мы получаем наужную нам информацию;

Основные действия со словарями

```python
contacts['Евгений'] = "+792307923850" # <- Таким образом можно добавить запись в словарь
```

```python
contacts.keys() # <- так можно получить  списко ключей, которые есть в словаре
```

```python
print(contacts.get('Денис')) # а вот еще один способ получения значения по ключу, только в этот раз если обратиться к 
#несуществующему ключу, вместо ошибки вернутся "ничего"
```

```python
print(contacts.values()) # так можно получить значения без ключей
```

```python
del contacts['Иван'] # так производится удаление по ключу
print(contacts)
```

```python
a = 4952304939
print(str(a))
```
